```python
%matplotlib widget
import os, sys
import matplotlib.pyplot as plt, pandas as pd, numpy as np
sys.path.append('../../../')
import default_style
```

```python
data = pd.read_csv('bpm.csv', delimiter=',', names=['time_ns', 'amplitude_au'])
fig, ax = plt.subplots(figsize=(9, 6))

x = 2.25*data.time_ns.rolling(2).mean()
y = data.amplitude_au.rolling(2).mean()

ax.plot(x, y, marker='+', linestyle='None', color='k')
ax.set_xlim(0, 110)
ax.set_ylim(0, 30)
ax.set_xlabel('Length [mm]')
ax.set_ylabel('Amplitude [a.u.]')

fig.tight_layout()

plt.savefig('/Users/alexisdevitre/Documents/GitHub/2025-devitre-PhD/chapter 3/3-3 Flux uniformity/raw/bpm.png')
```
