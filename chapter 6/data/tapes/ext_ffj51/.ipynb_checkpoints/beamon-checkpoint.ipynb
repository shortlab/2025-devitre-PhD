```python
def estimate_step_dose(data, t0, t1, t2, t3, vb=False):
    d = data[((t0 <= data.time_s) & (data.time_s < t1)) | ((t2 <= data.time_s) & (data.time_s < t3))]
    x = np.arange(d.time_s.min(), d.time_s.max(), 1)
    y = interp1d(d.time_s, d.ibeam_nA)(x)
    if vb:
        fig, ax = plt.subplots()
        ax.plot(x, y, marker='o', linestyle='None')
        ax.plot(d.time_s, d.ibeam_nA, marker='+', color='k')
    return np.trapz(y, dx=1)
```
