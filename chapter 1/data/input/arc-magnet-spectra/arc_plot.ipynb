```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import LogLocator

# Generate data
x = np.logspace(0.1, 2, 100)  # x from 10^0.1 to 10^2
y = x**2

# Create a plot with log scale
fig, ax = plt.subplots()
ax.plot(x, y, label="y = x^2")

# Set log scale on both axes
ax.set_xscale('log')
ax.set_yscale('log')

# Configure major and minor ticks
ax.xaxis.set_major_locator(LogLocator(base=10.0, numticks=10))  # One major tick per order of magnitude
ax.xaxis.set_minor_locator(LogLocator(base=10.0, subs=np.arange(1.0, 10) / 10, numticks=100))  # Minor ticks
ax.yaxis.set_major_locator(LogLocator(base=10.0, numticks=10))
ax.yaxis.set_minor_locator(LogLocator(base=10.0, subs=np.arange(1.0, 10) / 10, numticks=100))

# Add gridlines
ax.grid(visible=True, which='major', color='black', linewidth=0.75)
ax.grid(visible=True, which='minor', color='gray', linestyle=':', linewidth=0.5)

# Add labels and title
ax.set_title("Log Scale with Custom Ticks")
ax.set_xlabel("X-axis (log scale)")
ax.set_ylabel("Y-axis (log scale)")

# Show legend and plot
ax.legend()
plt.show()
```
